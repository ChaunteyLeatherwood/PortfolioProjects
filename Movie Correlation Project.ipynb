```python
# Import packages for project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Read in the data
df = pd.read_csv(r'C:\Users\User\Documents\Data Analysis Portfolio\Correlation in Python\movies.csv.zip')
```

```python
df.head()
```

```python
# Check for missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Replace null values with 0

df.replace(np.nan, 0, inplace = True)
```

```python
# Data Types for columns

print(df.dtypes)
```

```python
# Change Data Types

df.budget = df.budget.astype('int64')
df.gross = df.gross.astype('int64')
df.votes = df.votes.astype('int64')
df.runtime = df.runtime.astype('int64')
```

```python
df
```

```python
print(df.dtypes)
```

```python
# Create correct Released column

df['releasedcorrected'] = df['released'].astype(str).str.split(' ').str[2]

df
```

```python
# Order by highest grossing movie

df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df
```

```python
# Drop any duplicates

df.company = df.company.astype(str)
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
# Plot Budget vs Gross using Seaborn

ax = sns.regplot(x="budget", y="gross", data=df,
                 scatter_kws={"color": "black"}, line_kws={"color": "red"})

plt.show()
```

```python
# Looking at Correlation

#df.corr()
df.corr(method='pearson')
#df.corr(method='kendall')
#df.corr(method='spearman')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns: 
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes  
        
df_numerized
```
